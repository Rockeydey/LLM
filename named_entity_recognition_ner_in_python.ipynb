# Import libraries

```python
import pandas as pd
import spacy
import requests
from bs4 import BeautifulSoup
from spacy import displacy # used to visualize the named entities in a text.
nlp = spacy.load("en_core_web_sm") #natural language processing library to process text and extract named entities
pd.set_option("display.max_rows", 200)
```

# Use case

```python
content = "This artical was published on 1st Nov, 2023. Trinamool Congress leader Mahua Moitra has moved the Supreme Court against her expulsion from the Lok Sabha over the cash-for-query allegations against her. Moitra was ousted from the Parliament last week after the Ethics Committee of the Lok Sabha found her guilty of jeopardising national security by sharing her parliamentary portal's login credentials with businessman Darshan Hiranandani."
```

```python
doc = nlp(content)
doc
```

```python
displacy.render(doc, style="ent",jupyter=True)
```

```python
for i in doc.ents:
  print(i.text, i.start_char, i.end_char, i.label_)
```

```python
# Above code can be written in list comprehension
entities = [(i.text, i.start_char, i.end_char, i.label_) for i in doc.ents]
entities
```

```python
df= pd.DataFrame(data=entities, columns=["text", "start_char", "end_char", "label_"])
df
```
